```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt

a = rand.randn(100) * 10
# Creates an array of 100 random numbers..
index = a > 0.2
# Array filter for numbers less than 0.2..
b = a[index]
# Writting changes to b..
c = np.sin(b)
# Numpy function for tangent of all elements in b..
plt.style.use('ggplot')
# Selecting plot style..
plt.bar(b,c);
# Plot..
```
