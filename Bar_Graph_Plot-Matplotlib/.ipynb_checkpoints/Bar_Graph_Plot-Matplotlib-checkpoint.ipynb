```python
import numpy as np   
import matplotlib.pyplot as plt

plt.style.use('ggplot')

y = np.arange(1,7)
y.flat[[0,1,2,3,4,5]] = 10,25,50,13,1,5

x = ['Unicycles','Bicycles','Motorcycles','Cars','Airplanes','Helicopters']

plt.bar(x,y);
```
